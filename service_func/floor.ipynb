```python
df['floor'] = df['floor'].fillna(0).astype('str')

def clean_floor(text):
    text = re.sub('[%s]' % re.escape("""!"#$%&'()*+/:;<=>?@[\]^_`{|}~"""), '', text)
    text = re.sub(r'подвал, 1', '1', text)
    text = re.sub(r'подвал', '-1', text)
    text = re.sub(r'цоколь, 1', '1', text)
    text = re.sub(r'1,2,антресоль', '1', text)
    text = re.sub(r'Подвал', '-1', text)
    text = re.sub(r'Цоколь', '-1', text)
    text = re.sub(r'цоколь', '-1', text)
    text = re.sub(r'тех.этаж 6', '6', text)
    text = re.sub(r'фактически на уровне 1 этажа', '1', text)
    text = re.sub(r'1,2,3', '1', text)
    text = re.sub(r'5, мансарда', '5', text)
    text = re.sub(r'мезонин', '1', text)
    text = re.sub(r'3, Мансарда (4 эт)', '3', text)
    text = re.sub(r'1, антресоль', '1', text)
    text = re.sub(r'мансарда (4эт)', '4', text)
    text = re.sub(r'мансарда', '2', text)
    text = re.sub(r'1, цоколь', '1', text)
    text = re.sub(r'3 антресоль', '3', text)
    text = re.sub(r'1 этаж', '1', text)
    text = re.sub(r'1-й', '1', text)
    text = re.sub(r'3 этаж', '3', text)
    text = re.sub(r'4 этаж', '4', text)
    text = re.sub(r'5 этаж', '5', text)
    text = re.sub(r'3, мансарда', '3', text)
    text = re.sub(r'123.0', '1', text)
    text = re.sub(r'1, -1', '1', text)
    text = re.sub(r'1,4', '1', text)
    text = re.sub(r'1,2', '1', text)
    text = re.sub(r'1,4,5', '1', text)
    text = re.sub(r'1-3', '1', text)
    text = re.sub(r'1 -1ный этаж', '1', text)
    text = re.sub(r'-1,1', '1', text)
    text = re.sub(r'1, 2.', '1', text)
    text = re.sub(r'-1 , 1 ', '1', text)
    text = re.sub(r'1, 2', '1', text)
    text = re.sub(r'1  -1 без отделки', '1', text)
    text = re.sub(r'2,3', '2', text)
    text = re.sub(r'4, 5', '4', text)
    text = re.sub(r'2,3', '2', text)
    text = re.sub(r'1, 2-й', '1', text)
    text = re.sub(r'1-7, техэтаж', '1', text)
    text = re.sub(r'1, 2, 3', '1', text)
    text = re.sub(r'-1, 1,22', '1', text)
    text = re.sub(r'-1, 3. 4', '3', text)
    text = re.sub(r'1-4', '1', text)
    text = re.sub(r'подва, 1.2 этаж', '1', text)
    text = re.sub(r'2, 3', '2', text)
    text = re.sub(r'1й', '1', text)
    text = re.sub(r'1 3', '2', text)
    text = re.sub(r'-1, 12', '-1', text)
    text = re.sub(r'1.2', '1', text)
    text = re.sub(r'7,8', '2', text)
    text = re.sub(r'-1, 1', '1', text)
    text = re.sub(r'3, 2', '2', text)
    text = re.sub(r'1,5', '1', text)
    text = re.sub(r'113', '11', text)
    text = re.sub(r'13,4,5,6', '13', text)
    text = re.sub(r' 1 Антресоль', '1', text)
    text = re.sub(r'2 4', '2', text)
    text = re.sub(r'-1ный', '-1', text)
    text = re.sub(r'3,4', '3', text)
    text = re.sub(r'1 и 4', '1', text)
    text = re.sub(r'технический этаж,5,6', '5', text)
    text = re.sub(r' 1, -1ьный', '1', text)
    text = re.sub(r'2, мансардный', '2', text)
    text = re.sub(r'1, технический этаж', '1', text)
    text = re.sub(r'3, 4', '3', text)
    text = re.sub(r'1и, -1 188,4 кв.м, -1 104 кв.м', '1', text)
    text = re.sub(r'1, -1', '1', text)
    text = re.sub(r'2-й', '2', text)
    text = re.sub(r'1этаж', '1', text)
    text = re.sub(r'1-7', '1', text)
    text = re.sub(r'1 по док-м -1', '1', text)
    text = re.sub(r'1,-1 ', '1', text)
    text = re.sub(r'2, 4, тех.этаж', '1', text)
    text = re.sub(r'-1ный, 1', '1', text)
    text = re.sub(r'Техническое подполье', '-1', text)
    text = re.sub(r' 1ьный', '1', text)
    text = re.sub(r'-1, 1', '1', text)
    return text

text = df['floor']
text = text.apply(clean_floor)
df['floor'] = text

df['floor'] = df['floor'].astype('float').astype('int')
```
